##Display the page source code

```
ipl_2019_url="https://www.iplt20.com/stats/2019"
r=requests.get(ipl_2019_url)
soup_2019=bs(r.content)
print(soup_2019.prettify())

```

```
table_standings=soup_2019.select('main#main-content div.stats-content table.standings-table')
print(table_standings)
```

```
table_headers=table_standings[0].select('th')
table_headers
```

## Columns for Standings 2019


```
table_columns=[]

for i in range(len(table_headers)):
    table_columns.append(table_headers[i].string)

table_columns[0]='Positions'
print(table_columns)
```

##Teams 

```
teams_list=table_standings[0].select('a[href*="/teams/"] [class*=team-name--short]')
teams_list
```

```
teams_short=[]
[teams_short.append(teams_list[i].string) for i in range(len(teams_list))]
teams_short
```

```
teams_full_list=table_standings[0].select('a[href*="/teams/"]')
teams_full_list
```

```
teams_full=[]
for i in range(len(teams_full_list)):
    first_element=teams_full_list[i].select('span:nth-of-type(1)')
    teams_full.append(first_element[0].get_text(strip=True))
teams_full
```

```
teams_names=dict()
for i in range(len(teams_short)):
    teams_names[teams_short[i]]=teams_full[i]
teams_names
```

##Rows for standings 2019

```
tr=table_standings[0].select('tr')
tr
```

```
l=[]

for rows in tr:
    data=[]
    row=rows.select('td')
    [data.append(row[i].get_text(strip=True)) for i in range(len(row))]
    l.append(data)

standings_df=pd.DataFrame(l, columns=table_columns)
standings_df=standings_df.drop(index=0, axis=0)
standings_df=standings_df.replace(to_replace=standings_df.Team.values,value=teams_full)
standings_df
```

```
qualified=[]
# [qualified.append('Q') for i in range(len(standings_df)) if standings_df.Positions.values[i]=='Q' ]
# qualified

for i in range(len(standings_df)):
    if standings_df.Positions.values[i]=='Q': 
        qualified.append('Qualified')
        standings_df.Positions.values[i]=i+1
    else:
        qualified.append('No')
qualified

standings_df['Playoffs']=qualified
standings_df
```
